In-line text to audio

```python
from gtts import gTTS #Google Text-to-Speech

#Add your own text
text="Greetings from Google! This is the sample text."
language='en'
output= gTTS(text=text, lang=language)
output.save("output.mp3")

from IPython.display import Audio #Package to show media onto the Colab/ Jupyter Notebook
Audio("output.mp3")
```

If the user wants to convert an uploaded .txt file into audio

```python
fileLocation = open("sample.txt", "r")
fileText = fileLocation.read().replace("\n", " ")
newOutput = gTTS(text=fileText, lang=language, slow=False)
newOutput.save("newOutput.mp3")

from IPython.display import Audio
Audio("newOutput.mp3")
```

Customizability to Speech in gTTS

```python
#for changing the accent of the language spoken use tld
output= gTTS(text=text, lang=language, tld='co.uk') #top level domain

#subtitute for the above to get the localized UK accent
```
